Task0:Build Python Function to check the given number prime or not

prompt:Write python function to check the given number or not

```python
def is_prime(num):
    """
    Checks if a given number is prime.

    Args:
        num: The number to check.

    Returns:
        True if the number is prime, False otherwise.
    """
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage:
print(is_prime(11))
print(is_prime(4))
```



if num <= 1:: This line checks if the input number num is less than or equal to 1. By definition, prime numbers are greater than 1, so if the number is 1 or less, the function immediately returns False.
`for i in range(2, int(num0.5) + 1):**: This is a loop that iterates through a range of numbers starting from 2 up to the square root ofnum(inclusive). We only need to check for divisors up to the square root of the number because if a numbernum` has a divisor greater than its square root, it must also have a divisor less than its square root.
if num % i == 0:: Inside the loop, this line checks if num is divisible by the current number i with no remainder. If it is, it means i is a divisor of num.
return False: If a divisor is found (i.e., num % i == 0 is true), the function immediately returns False because a prime number is only divisible by 1 and itself.
return True: If the loop completes without finding any divisors (meaning the number is not divisible by any number from 2 up to its square root), the function returns True, indicating that the number is prime.
